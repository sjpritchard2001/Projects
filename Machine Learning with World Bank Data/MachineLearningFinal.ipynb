# Research Question

In this project we sought to answer the following key questions:
1. What is the impact of key variables such as education, gdp growth rate, technology imports and exports on unemployment rates?
2. Do unemployment rates respond to election cycles?

# Methodology and Model Selection
 

We employed multiple linear regression techniques, utilizing panel data models to explore the impact of key drivers on unemployment rates. Specifically, we worked with the pooled OLS model, fixed effects model, and random effects model. Upon evaluating the models, we observed that their error metrics—mean absolute error (MAE), mean absolute percentage error (MAPE), and mean squared error (MSE)—were closely aligned. Ultimately, we chose to rely on the estimates of the random effects model as it demonstrated the highest adjusted R-squared value, and the results of the Hausman test supported the appropriateness of this model.

# Key Results

The regression analysis revealed several key factors significantly influencing unemployment rates in developed countries. Among these, inflation, GDP growth, population, ICT imports, and tertiary enrollment emerged as significant predictors. Notably, with the exception of tertiary enrollment, the coefficients for inflation, GDP growth, population, and ICT imports were negative, suggesting a generally inverse relationship with unemployment rates.

**Explaining the Findings:**
**Inflation:**
The negative relationship between inflation and unemployment could reflect a manifestation of the Phillips Curve, which postulates an inverse relationship between the two variables in the short term. In developed economies, mild inflation often correlates with robust economic activity, higher demand for labor, and, consequently, lower unemployment rates.

**GDP Growth:**
The negative coefficient for GDP growth aligns with economic theory; as the economy grows, businesses expand, leading to increased labor demand and reduced unemployment. This finding underscores the importance of sustained economic growth in mitigating unemployment in developed countries.

**Population:**
The negative association between population and unemployment may be explained by the structure of labor markets in developed countries. These nations often have well-established mechanisms for labor absorption, particularly for growing populations, such as policies encouraging workforce participation and immigration targeted at addressing skill gaps.

**ICT Imports:**
The negative impact of ICT imports on unemployment could indicate that increased access to advanced technology facilitates economic productivity and job creation in sectors reliant on technology. In developed countries, imports of ICT products might stimulate innovation, efficiency, and competitive industries, which could support employment.

**Tertiary Enrollment:**
Unlike the other variables, tertiary enrollment showed a significant positive coefficient. This might reflect the phenomenon of more individuals pursuing higher education delaying their entry into the labor market, thereby temporarily increasing the unemployment rate among the population. Alternatively, it could indicate a mismatch between the skills provided by higher education institutions and the demands of the job market in these countries.

**Election Cycles:**
The dummy variable representing election cycles did not have a statistically significant impact on unemployment rates. However, the negative coefficient, though not significant, is worth noting. It might reflect efforts by governments to stimulate the economy and reduce unemployment during election years, such as through expansionary fiscal policies or other interventions aimed at garnering public approval.

**Implications of the Findings:**
These results highlight the multifaceted nature of unemployment in developed countries, influenced by a combination of macroeconomic factors, technological adoption, demographic shifts, and education systems. Policymakers should consider these factors when designing strategies to reduce unemployment, balancing short-term economic interventions with long-term structural reforms. Furthermore, the insignificant yet negative coefficient for election cycles might suggest potential political dynamics at play, which could warrant further investigation.

```r
# Identify outliers using the IQR method
identify_outliers <- function(dataUnemp) {
  outlier_info <- data.frame(Variable = character(), Outliers = numeric(), stringsAsFactors = FALSE)
  
  # Loop through all numeric columns except 'Year'
  for (col in names(dataUnemp)) {
    if (is.numeric(dataUnemp[[col]]) && col != "Year") {
      
      # Calculate IQR
      Q1 <- quantile(dataUnemp[[col]], 0.25, na.rm = TRUE)
      Q3 <- quantile(dataUnemp[[col]], 0.75, na.rm = TRUE)
      IQR_value <- IQR(dataUnemp[[col]], na.rm = TRUE)
      
      # Define outlier boundaries
      lower_bound <- Q1 - 1.5 * IQR_value
      upper_bound <- Q3 + 1.5 * IQR_value
      
      # Find outliers
      outliers <- sum(dataUnemp[[col]] < lower_bound | dataUnemp[[col]] > upper_bound, na.rm = TRUE)
      
      # Store outliers information
      outlier_info <- rbind(outlier_info, data.frame(Variable = col, Outliers = outliers))
    }
  }
  
  return(outlier_info)
}

# Apply the function to your dataset
outliers_summary <- identify_outliers(dataUnemp)

# Print the summary of outliers for each variable
print(outliers_summary)
```

```r
# Function to identify outliers per country and variable, and return detailed info
identify_outliers_by_country <- function(dataUnemp, id_vars = c("Country_Code", "Year")) {
  # Ensure the dataset includes necessary identifying columns
  if (!all(id_vars %in% colnames(dataUnemp))) {
    stop("Ensure 'Country_Code' and 'Year' (or other identifiers) are in the dataset.")
  }
  
  # Initialize a data frame to store detailed outlier info
  outlier_details <- data.frame(
    Country_Code = character(),
    Year = numeric(),
    Variable = character(),
    Value = numeric(),
    stringsAsFactors = FALSE
  )
  
  # Loop through all numeric variables
  numeric_cols <- setdiff(names(dataUnemp), id_vars)  # Exclude identifying variables
  for (col in numeric_cols) {
    if (is.numeric(dataUnemp[[col]])) {
      # Calculate IQR for the column
      Q1 <- quantile(dataUnemp[[col]], 0.25, na.rm = TRUE)
      Q3 <- quantile(dataUnemp[[col]], 0.75, na.rm = TRUE)
      IQR_value <- Q3 - Q1
      
      # Define outlier thresholds
      lower_bound <- Q1 - 1.5 * IQR_value
      upper_bound <- Q3 + 1.5 * IQR_value
      
      # Identify rows that contain outliers for the current variable
      outlier_rows <- which(dataUnemp[[col]] < lower_bound | dataUnemp[[col]] > upper_bound)
      
      # If outliers exist, store their information
      if (length(outlier_rows) > 0) {
        for (row in outlier_rows) {
          outlier_details <- rbind(
            outlier_details,
            data.frame(
              Country_Code = as.character(dataUnemp[row, "Country_Code"]),
              Year = as.numeric(dataUnemp[row, "Year"]),
              Variable = as.character(col),
              Value = as.numeric(dataUnemp[row, col]),
              stringsAsFactors = FALSE
            )
          )
        }
      }
    }
  }
  
  return(outlier_details)
}

# Apply the function to your dataset
outliers_detailed <- identify_outliers_by_country(dataUnemp)

# View the first few rows of the detailed outliers
head(outliers_detailed)

# Print summary statistics for outliers by variable and country
summary_outliers <- aggregate(
  Value ~ Country_Code + Variable,
  data = outliers_detailed,
  FUN = length
)
colnames(summary_outliers)[3] <- "Number_of_Outliers"
print(summary_outliers)

# Save the detailed outlier information to a CSV (if needed)
write.csv(outliers_detailed, "detailed_outliers.csv", row.names = FALSE)
```

# Fitting additional panel regression models: Fixed Effects and Random Effects Models

**In this section we run the fixed effects and random effects model and test for heteroskedasticity and autocorrelation. Given that the residuals are heteroskedastic, also go ahead to re-run the models to using estimators that are robust to both heteroskedasticity and autocorrelation**

```r
install.packages("plm")
install.packages("lmtest") # Add this line to install the lmtest package
library(plm)
library(lmtest) # Add this line to load the lmtest package

# Fixed Effects model
fe_model <- plm(Unemployment ~ inflation + GDP_Growth + population + ICT_Imports + Primary_Enrollment + Secondary_Enrollment + Tertiary_Enrollment + election_year,data = trainData, model = "within")

summary(fe_model)

# Random Effects model
re_model <- plm(Unemployment ~ inflation + GDP_Growth + population + ICT_Imports + Primary_Enrollment + Secondary_Enrollment + Tertiary_Enrollment + election_year,data = trainData, model = "random")

summary(re_model)

# Perform the Hausman Test to determine the best model (FE vs RE)
hausman_test <- phtest(fe_model, re_model)
print(hausman_test)

# Breusch-Pagan test for heteroskedasticity
bp_test <- bptest(re_model)
bp_test

# Wooldridge test for autocorrelation
wooldridge_test <- pbgtest(re_model)
wooldridge_test



# Using the selected variables from the stepwise regression run previously (model1)
re_modelstep <- plm(Unemployment ~ inflation + ICT_Imports + Primary_Enrollment + 
    Secondary_Enrollment + Tertiary_Enrollment,data = trainData, model = "random")


#summary(re_modelstep)
```

```r
# correcting for autocorrelation for re_modelstep
# Now compute robust standard errors that are corrected for autocorrelation
# vcovHC() function calculates heteroskedasticity-consistent standard errors
# type = "HC1" for White standard errors (robust to heteroskedasticity)
# You can specify a lag (type = "HC3" for more robust options) or cluster by country
robust_se <- vcovHC(re_modelstep, type = "HC1", cluster = "group")

# Display the summary with robust standard errors
coeftest(re_modelstep, vcov. = robust_se)
```

```r
# correcting for autocorrelation for re_model
robust_se <- vcovHC(re_model, type = "HC1", cluster = "group")

# Display the summary with robust standard errors
coeftest(re_model, vcov. = robust_se)
```

# Model Evaluation: model1, model2, fe_model, re_model, re_modelstep

In this section, we evaluate the effectiveness of the model for predictions drawing on the MAE,RMSE and MAPE. We also use the K-fold cross validation approach. The evaluation metrics are very close for each of the models and opted to use the estimates of the re-model based on the hausman test run in the previous section and given that this model has the highest adjusted r-squared.  

```r
# Load necessary libraries
install.packages("Metrics")  # If not already installed
library(Metrics)

# Define a function to calculate error metrics
calculate_metrics <- function(model, testData, y_actual) {
  # Predict using the model
  y_pred <- predict(model, newdata = testData)
  
  # Calculate error metrics
  mae <- mae(y_actual, y_pred)  # Mean Absolute Error
  rmse <- rmse(y_actual, y_pred)  # Root Mean Squared Error
  mape <- mean(abs((y_actual - y_pred) / y_actual)) * 100  # Mean Absolute Percentage Error
  
  # Return the results as a list
  return(list(MAE = mae, RMSE = rmse, MAPE = mape))
}

# Get the actual 'Unemployment' values from the test data
y_actual <- testData$Unemployment

# Apply the function to each model
metrics_model1 <- calculate_metrics(model1, testData, y_actual)
metrics_model2 <- calculate_metrics(model2, testData, y_actual)
metrics_re_model <- calculate_metrics(re_model, testData, y_actual)
metrics_re_modelstep <- calculate_metrics(re_modelstep, testData, y_actual)
metrics_fe_model <- calculate_metrics(fe_model, testData, y_actual)

# Print the results
cat("Error Metrics for Model 1:\n")
print(metrics_model1)

cat("\nError Metrics for Model 2:\n")
print(metrics_model2)

cat("\nError Metrics for Random Effects Model:\n")
print(metrics_re_model)

cat("\nError Metrics for Stepwise Model:\n")
print(metrics_re_modelstep)


cat("\nError Metrics for fixed Effects Model:\n")
print(metrics_fe_model)
```

```r
# Load necessary libraries
install.packages("caret")  # If not already installed
library(caret)
install.packages("Metrics")  # If not already installed
library(Metrics)

# Define a function to calculate error metrics
calculate_metrics <- function(y_actual, y_pred) {
  mae_value <- mae(y_actual, y_pred)  # Mean Absolute Error
  rmse_value <- rmse(y_actual, y_pred)  # Root Mean Squared Error
  mape_value <- mean(abs((y_actual - y_pred) / y_actual), na.rm = TRUE) * 100  # Mean Absolute Percentage Error
  return(list(MAE = mae_value, RMSE = rmse_value, MAPE = mape_value))
}

# Define a function to perform k-fold cross-validation
k_fold_cv <- function(model, data, k) {
  # Remove rows with NA values in the target variable
  data <- na.omit(data)
  
  # Create k-folds
  folds <- createFolds(data$Unemployment, k = k, list = TRUE, returnTrain = TRUE)
  
  # Calculate metrics for each fold
  metrics_list <- lapply(folds, function(train_indices) {
    train_data <- data[train_indices, ]
    test_data <- data[-train_indices, ]
    y_actual <- test_data$Unemployment
    y_pred <- predict(model, newdata = test_data)
    calculate_metrics(y_actual, y_pred)
  })
  
  # Combine metrics for all folds
  metrics_df <- do.call(rbind, lapply(metrics_list, as.data.frame))
  
  # Return the average of the metrics for each model
  return(colMeans(metrics_df, na.rm = TRUE))
}

# Set the number of folds
k <- 5

# Perform k-fold cross-validation for each model
cv_metrics_model1 <- k_fold_cv(model1, dataUnemp1, k)
cv_metrics_model2 <- k_fold_cv(model2, dataUnemp1, k)
cv_metrics_re_model <- k_fold_cv(re_model, dataUnemp1, k)
cv_metrics_re_modelstep <- k_fold_cv(re_modelstep, dataUnemp1, k)

# Report the results
print("CV Metrics for Model 1:")
print(cv_metrics_model1)

print("CV Metrics for Model 2:")
print(cv_metrics_model2)

print("CV Metrics for Random Effects Model:")
print(cv_metrics_re_model)

print("CV Metrics for Stepwise Random Effects Model:")
print(cv_metrics_re_modelstep)
```
